**Net Present Value (NPV)**

#Use an 8% discount rate.

```python
# Set the discount rate (8%)
discount_rate = 0.08

# Create Year Index (0, 1, 2, ..., 9)
df["Year_Index"] = range(len(df))

# Calculate Discount Factor
df["Discount_Factor"] = 1 / ((1 + discount_rate) ** df["Year_Index"])

# Calculate Discounted Net Benefit
df["Discounted_Net_Benefit"] = (
    df["Net_Benefit"] * df["Discount_Factor"]
)

# Calculate Net Present Value (NPV)
NPV = df["Discounted_Net_Benefit"].sum()

print(f"Net Present Value (NPV): {NPV:.2f} Billion KES")

# View the results
df[["Year", "Net_Benefit", "Discount_Factor", "Discounted_Net_Benefit"]]
```

**INSIGHTS**

The Net Present Value (NPV) analysis evaluates the economic viability of the Nairobi Expressway by accounting for the time value of money. Using a discount rate of 8%, future net benefits are converted into their present values through the application of a discount factor. As expected, the discount factor declines over time because benefits received in the future are worth less than benefits received today. Consequently, while the project continues to generate increasing net benefits, their present values decrease slightly when discounted.

The discounted net benefits remain positive throughout the operational years of the project, indicating that the long-term economic gains are sufficient to offset both the initial construction costs and ongoing operational expenses. The calculated Net Present Value (NPV) is positive, demonstrating that the present value of the project's benefits exceeds the present value of its costs. This is a strong indicator that the Nairobi Expressway creates economic value rather than destroying it.

Overall, the positive NPV confirms that the Nairobi Expressway is a financially and economically viable investment under the Public-Private Partnership (PPP) model. The findings suggest that the project delivers sustainable long-term returns, improves transport efficiency, enhances productivity, and contributes positively to Kenya's economic development. A positive NPV also supports the use of PPP arrangements for financing major infrastructure projects, as they can generate significant value for both investors and the public over the project's lifecycle.

**Policy Recommendations to the Government of Kenya**
    

Based on the findings of the Cost-Benefit Analysis, the Nairobi Expressway has demonstrated strong economic viability and has the potential to contribute even more to Kenya's long-term development. The following recommendations are proposed to maximize its contribution to GDP growth, employment creation, industrialization, and national competitiveness.


1. Expand the Nairobi Expressway into a National Transport Corridor

The Government should extend the expressway network to connect Jomo Kenyatta International Airport (JKIA), the Nairobi Central Business District, the Nairobi Inland Container Depot, the Standard Gauge Railway (SGR), and major industrial zones. Improved connectivity will reduce logistics costs, facilitate trade, and stimulate investment, thereby contributing directly to GDP growth.

    
2. Establish Industrial and Logistics Parks Along the Corridor

Special Economic Zones (SEZs), logistics hubs, and industrial parks should be developed along the expressway corridor. These facilities can attract manufacturing firms, warehouses, distribution centers, and technology companies, creating thousands of direct and indirect employment opportunities while expanding Kenya's export capacity.


3. Promote Local Employment and Skills Development

Future infrastructure projects should prioritize the employment of Kenyan engineers, technicians, artisans, data analysts, environmental specialists, and project managers. Partnerships between government, universities, and the private sector should strengthen technical training and apprenticeship programs to equip young people with market-relevant skills.

    
4. Strengthen Public-Private Partnerships (PPP)

The Government should continue using the PPP model for economically viable infrastructure projects while ensuring transparent procurement, effective contract management, and fair risk sharing. Well-managed PPPs can mobilize private investment without placing excessive pressure on public finances.

    
5. Adopt Gemba Kaizen Across Infrastructure Operations

Gemba Kaizen should be integrated into the management of the Nairobi Expressway and related transport infrastructure. Continuous improvement through regular inspections, data-driven decision-making, preventive maintenance, and employee involvement can improve operational efficiency, reduce maintenance costs, minimize traffic disruptions, and enhance service quality for road users.

    
6. Apply Ikigai to Workforce Development

The Government and private operators should embrace Ikigai as a guiding principle for workforce development. By aligning employees' skills, passion, purpose, and organizational needs, institutions can improve productivity, innovation, employee satisfaction, and long-term retention. A motivated workforce is essential for sustainable infrastructure management and economic growth.


7. Accelerate Digital Transformation

Expand intelligent transport systems (ITS), electronic toll collection, artificial intelligence for traffic management, and real-time monitoring systems. Digital technologies will improve traffic flow, reduce congestion, enhance safety, and generate valuable data for infrastructure planning and policy formulation.

    
8. Encourage Green and Sustainable Transport

Invest in electric vehicle charging infrastructure, environmentally friendly public transport, green landscaping, and renewable energy for expressway operations. Sustainable transport initiatives can reduce emissions while supporting Kenya's climate commitments and improving environmental quality.


9. Support Small and Medium Enterprises (SMEs)

Create business opportunities for local entrepreneurs through roadside service centres, logistics support services, vehicle maintenance facilities, hospitality businesses, and digital service providers along the expressway corridor. Supporting SMEs will stimulate entrepreneurship, increase household incomes, and generate additional employment.

    
10. Institutionalize Data-Driven Decision-Making

Government agencies should establish integrated data platforms to continuously monitor traffic volumes, toll revenues, maintenance performance, environmental impacts, accident rates, and customer satisfaction. Evidence-based policy decisions will improve infrastructure planning and maximize the economic returns of future investments.
